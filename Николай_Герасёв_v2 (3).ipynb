–≠–∫—Å–ø–µ—Ä—Ç–∞ 3, –Ω–æ –æ—Ü–µ–Ω–∫–∞ –Ω—É–∂–Ω–∞ –æ–¥–Ω–∞, –ø–æ—ç—Ç–æ–º—É –∏—Å–ø–æ–ª—å–∑—É–µ–º –º–µ—Ç–æ–¥ –≥–æ–ª–æ—Å–æ–≤–∞–Ω–∏—è: –µ—Å–ª–∏ 2 —ç–∫—Å–ø–µ—Ä—Ç–∞ –ø–æ—Å—Ç–∞–≤–∏–ª–∏ –æ–¥–∏–Ω–∞–∫–æ–≤—É—é –æ—Ü–µ–Ω–∫—É, –æ–Ω–∞ —Å—á–∏—Ç–∞–µ—Ç—Å—è –≤–µ—Ä–Ω–æ–π. –µ—Å–ª–∏ –≤—Å–µ —Ç—Ä–∏ –æ—Ü–µ–Ω–∫–∏ —Ä–∞–∑–Ω—ã–µ, —Ç–∞–∫–æ–µ –æ–ø–∏—Å–∞–Ω–∏–µ –æ—Ç–±—Ä–∞—Å—ã–≤–∞–µ—Ç—Å—è.

````python
def voting(v1, v2, v3):
    """—Ñ—É–Ω–∫—Ü–∏—è –ø—Ä–∏–Ω–∏–º–∞–µ—Ç —Ç—Ä–æ–π–∫—É –æ—Ü–µ–Ω–æ–∫ –∏ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç —Ç—É, —á—Ç–æ —Å–æ–≤–ø–∞–¥–∞–µ—Ç —Ö–æ—Ç—è –±—ã —É –¥–≤–æ–∏—Ö, –∏–Ω–∞—á–µ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç None"""
    if v1 == v2 or v1 == v3:
        return v1
    elif v2 == v3:
        return v2
    else:
        return None
````

````python
lemmatizer = WordNetLemmatizer()
````

–î–ª—è –¥–∞–ª—å–Ω–µ–π—à–µ–π —Ä–∞–±–æ—Ç—ã —Ç–µ–∫—Å—Ç—ã –Ω—É–∂–Ω–æ –ª–µ–º–º–∞—Ç–∏–∑–∏—Ä–æ–≤–∞—Ç—å –∏ —Ç–æ–∫–µ–Ω–∏–∑–∏—Ä–æ–≤–∞—Ç—å. –í —Ä—è–¥–µ —Å–ª—É—á–∞–µ–≤ –¥–ª—è –ø—Ä–∞–≤–∏–ª—å–Ω–æ–π –ª–µ–º–º–∞—Ç–∏–∑–∞—Ü–∏–∏ –ª–µ–º–º–∞—Ç–µ–∑–∞—Ç–æ—Ä—É —Ç—Ä–µ–±—É–µ—Ç—Å—è —É—Ç–æ—á–Ω–µ–Ω–∏–µ, –∫–∞–∫–æ–π —á–∞—Å—Ç—å—é —Ä–µ—á–∏ —è–≤–ª—è–µ—Ç—Å—è —Å–ª–æ–≤–∞ (POS). –≠—Ç—É –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é —Ç–∞–∫–∂–µ –º–æ–∂–Ω–æ –ø–æ–ª—É—á–∏—Ç—å —Å—Ä–µ–¥—Å—Ç–≤–∞–º–∏ nltk 

````python
def pos_tag(word):
    tag = nltk.pos_tag([word])[0][1][0].upper()
    tag_dict = {
        "J": wordnet.ADJ,
        "N": wordnet.NOUN,
        "V": wordnet.VERB,
        "R": wordnet.ADV,
    }
    # –µ—Å–ª–∏ —Å–ª–æ–≤–æ –Ω–µ —è–≤–ª—è–µ—Ç—Å—è —Å—É—â–µ—Å—Ç–≤–∏—Ç–µ–ª—å–Ω—ã–º, –ø—Ä–∏–ª–∞–≥–∞—Ç–µ–ª—å–Ω—ã–º, –≥–∞–≥–æ–ª–æ–º –∏–ª–∏ –Ω–∞—Ä–µ—á–∏–µ–º, –Ω–∞–∑–Ω–∞—á–∞–µ–º –µ–≥–æ —Å—É—â–µ—Å—Ç–≤–∏—Ç–µ–ª—å–Ω—ã–º
    return tag_dict.get(tag, wordnet.NOUN)
````

````python
%%time
train_dataset["lemm_query"] = train_dataset["query_text"].apply(
    lambda x: " ".join(
        [lemmatizer.lemmatize(w, pos_tag(w)) for w in nltk.word_tokenize(x)]
    )
)
````

````python
train_dataset["lemm_query"] = train_dataset["lemm_query"].apply(
    lambda x: (" ".join(re.sub(r"[^a-zA-Z ]", " ", x).split()).lower())
)
````

````python
def child_in_text(text: str) -> bool:
    """–ü–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ –∏—â–µ–º —Å–ª–æ–≤–∞ –∏–∑ —Å—Ç–æ–ø-–ª–∏—Å—Ç–∞ child_stop –≤ –∑–∞–¥–∞–Ω–Ω–æ–º —Ç–µ–∫—Å—Ç–µ text. –í–æ–∑–≤—Ä–∞—â–∞–µ–º True, –µ—Å–ª–∏ –Ω–∞—à–ª–æ—Å—å —Ö–æ—Ç—è –±—ã –æ–¥–Ω–æ —Å–ª–æ–≤–æ"""
    res = False
    for w in child_stop:
        res |= text.lower().find(w) > -1
    return res
````

````python
preprocess = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])
````

````python
# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏—è –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è –≤ —Ç–µ–Ω–∑–æ—Ä –∏ –ø–æ–ª—É—á–µ–Ω–∏—è —ç–º–±–µ–¥–¥–∏–Ω–≥–∞
def image_to_embedding(file: str, path: str):
    img_path = join(path, file)
    img = Image.open(img_path).convert('RGB')
    img_tensor = preprocess(img).unsqueeze(0)  # –î–æ–±–∞–≤–ª—è–µ–º —Ä–∞–∑–º–µ—Ä –±–∞—Ç—á–∞
    with torch.no_grad():
        embedding = resnet(img_tensor).flatten().numpy()
    return embedding
````

````python
class LinearRegressionModel(nn.Module):
    def __init__(self, input_dim, output_dim):
        super(LinearRegressionModel, self).__init__()
        self.linear = nn.Linear(input_dim, output_dim)

    def forward(self, x):
        return self.linear(x)
````

### –ù–µ–π—Ä–æ–Ω–Ω–∞—è —Å–µ—Ç—å

–¢–µ–ø–µ—Ä—å –æ–±—É—á–∏–º —Å—Ä–∞–≤–Ω–∏—Ç–µ–ª—å–Ω–æ –ø—Ä–æ—Å—Ç—É—é –Ω–µ–π—Ä–æ—Å–µ—Ç—å.

–ù–µ–π—Ä–æ—Å–µ—Ç—å –Ω–µ –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –Ω–∞ –≤—Ö–æ–¥ —Ç—Ä—ë—Ö–º–µ—Ä–Ω—ã–µ —Ç–µ–Ω–∑–æ—Ä—ã —Å —ç–ª–µ–º–µ–Ω—Ç–∞–º–∏ —Ç–∏–ø–∞ float64, —Ç–∞–∫ —á—Ç–æ –ø—Ä–µ–æ–±—Ä–∞–∑—É–µ–º –≤—Ö–æ–¥–Ω—ã–µ –∏ –≤—ã—Ö–æ–¥–Ω—ã–µ –≤–µ–∫—Ç–æ—Ä–∞ –≤ –¥–≤—É–º–µ—Ä–Ω—ã–µ —Ç–µ–Ω–∑–æ—Ä—ã float32

````python
X_nn = X.view(X.shape[0], -1)
Y_nn = Y.view(Y.shape[0], -1)

X_nn_val = X_val.view(X_val.shape[0], -1)
Y_nn_val = torch.tensor(np.stack(y_test["image_vector"].values), dtype=torch.float32).view(-1, image_vector_size)
````

````python
class NNModel(nn.Module):
    def __init__(self, input_dim, output_dim, dout=0.5):
        super(NNModel, self).__init__()
        self.fc1 = nn.Linear(input_dim, input_dim)
        self.fc2 = nn.Linear(input_dim, input_dim // 2)
        self.fc3 = nn.Linear(input_dim // 2, input_dim // 4)
        self.fc4 = nn.Linear(input_dim // 4, output_dim)
        self.dropout = nn.Dropout(dout)

    def forward(self, x):
        x = torch.relu(self.fc1(x))
        x = self.dropout(x)
        x = torch.relu(self.fc2(x))
        x = self.dropout(x)
        x = torch.relu(self.fc3(x))
        x = self.fc4(x)
        return x

model = NNModel(text_vector_size, image_vector_size, dout=0.5)
````

````python
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.01)
````

````python
epochs = 100
batch_size = 100

train_dataset = torch.utils.data.TensorDataset(X_nn, Y_nn)
train_loader = torch.utils.data.DataLoader(dataset=train_dataset, batch_size=batch_size, shuffle=True)

if isfile("./nn_model.pth"):
    model.load_state_dict(torch.load("./nn_model.pth"))

if not isfile("./nn_model.pth"):
    for epoch in range(epochs):
        model.train()
        for i, (inputs, targets) in enumerate(train_loader):
            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, targets)
            loss.backward()
            optimizer.step()

        if (epoch + 1) % 5 == 0:
            model.eval()
            val_outputs = model(X_nn_val)
            val_loss = criterion(val_outputs, Y_nn_val)
            print("Epoch [{}/{}], Loss: {:.4f}, Val Loss: {:.4f}".format(epoch+1, epochs, loss.item(), val_loss.item()))
            
    torch.save(model.state_dict(), "./nn_model.pth")
````

````python
def vec_distance(A, B, ord='euclidean'):
    return torch.norm(B - A, p=2).item()

def find_image_NN(query_vector, ivecs, lim=3):
    model.eval()
    with torch.no_grad():
        iv = model(query_vector)
        df = pd.DataFrame(ivecs)
        df['dist'] = df.apply(lambda row: vec_distance(iv, torch.tensor(row['image_vector'], dtype=torch.float32)), axis=1)
    return df.sort_values(by="dist")["image"].unique()[:lim]
````

````python
def show_images(query, images, path='train_images', rows=1, columns=3):
    fig = plt.figure(figsize=(12, 12))
    print(query)
    count = 1
    for file in images:
        img = Image.open(join(path, file))
        fig.add_subplot(rows, columns, count)
        count += 1
        plt.imshow(np.array(img))
        plt.axis("off")

def predict_NN(idx, df=y_test):
    """–°–æ–±–∏—Ä–∞–µ—Ç –Ω–µ—Å–∫–æ–ª—å–∫–æ –ø—Ä–µ–¥—ã–¥—É—â–∏—Ö —Ñ—É–Ω–∫—Ü–∏–π –≤ –æ–¥–Ω—É –¥–ª—è –ø—Ä–æ–≥–Ω–æ–∑–∏—Ä–æ–≤–∞–Ω–∏—è –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π —Å –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ–º –Ω–µ–π—Ä–æ—Å–µ—Ç–∏"""
    qv = torch.tensor(X_test.loc[idx, "query_vector"], dtype=torch.float32).view(1, -1)
    qt = test_df[test_df["annotation_id"] == X_test.loc[idx, "annotation_id"]]["query_text"].iloc[0]
    images = find_image_NN(qv, df)
    show_images(qt, images)
````
